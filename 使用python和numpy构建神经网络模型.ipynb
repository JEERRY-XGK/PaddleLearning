```python
import numpy
import json
import paddle
import matplotlib.pyplot as plt
```

```python
def load_data(path: str):
    feature_names = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']#   特征名称
    data = numpy.fromfile(file=path, sep=' ')#  读取数据
    data = data.reshape(len(data)//len(feature_names),len(feature_names))
    train_set = numpy.array(data[0:int(data.shape[0]*0.8)])
    maximum,minimum,average = train_set.max(axis=0),train_set.min(axis=0),train_set.sum(axis=0)/train_set.shape[0] #    求样本最大值，最小值，均值
    #   数据归一化
    for i in range(len(feature_names)):
        data[:,i] = (data[:,i]-average[i])/(maximum[i]-minimum[i])
    #   划分数据集
    train_set = numpy.array(data[:int(data.shape[0]*0.8)])
    test_set = numpy.array(data[int(data.shape[0]*0.8):])
    return train_set,test_set
```

```python
train_data,test_data = load_data(path='./data/housing.data')
```
